```python
def display_student_grades(students, student_name):
    """
    Displays individual quiz grades and the average for the chosen student.

    Args:
        students (dict): A dictionary of student names and their quiz grades.
        student_name (str): The name of the student to display grades for.
    """

    if student_name not in students:
        print(f"Student '{student_name}' not found.")
        return

    grades = students[student_name]
    quiz_count = len(grades)

    print(f"\nIndividual Quiz Grades for {student_name}:")
    for i in range(quiz_count):
        print(f"Quiz {i+1}: {grades[i]}")

    average = sum(grades) / quiz_count
    print(f"Average: {average:.2f}")

def calculate_and_display_averages(students):
    """
    Calculates and displays individual student averages, quiz averages, and class average.

    Args:
        students (dict): A dictionary of student names and their quiz grades.
    """

    # Calculate individual student averages
    student_averages = {student: sum(grades) / len(grades) for student, grades in students.items()}

    # Calculate quiz averages
    quiz_averages = {}
    for quiz_num in range(len(students["Keanu"])):  # Use any student's list as reference
        quiz_scores = [student[quiz_num] for student in students.values()]
        quiz_average = sum(quiz_scores) / len(quiz_scores)
        quiz_averages[quiz_num + 1] = quiz_average

    # Calculate class average
    class_average = sum([sum(grades) for grades in students.values()]) / (len(students) * len(grades))

    # Display all averages
    print("\nStudent Averages:")
    for student, average in student_averages.items():
        print(f"{student}: {average:.2f}")

    print("\nQuiz Averages:")
    for quiz_num, average in quiz_averages.items():
        print(f"Quiz {quiz_num}: {average:.2f}")

    print("\nClass Average: {:.2f}".format(class_average))

def main():
    # Sample student data
    students = {
        "Keanu": [80, 75, 90],
        "Erica": [95, 80, 85],
        "Tiffany": [70, 85, 92],
        "Lorenzo": [88, 90, 95],
        #"Eve": [72, 82, 88],
    }

    while True:
        print("\nChoose an option:")
        print("1. View individual student grades")
        print("2. View overall class averages")
        print("3. Exit")

        choice = input("Enter your choice (1, 2, or 3): ")

        if choice == "1":
            student_name = input("Enter the student's name: ").capitalize()
            display_student_grades(students, student_name)
        elif choice == "2":
            calculate_and_display_averages(students)
        elif choice == "3":
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please enter 1, 2, or 3.")

if __name__ == "__main__":
    main()
```
